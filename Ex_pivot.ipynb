# Pivot 
#### pivot allows to transform or reshape data


```python
import pandas as pd 
df = pd.read_csv('weather.csv')
```

```python
# df.pivot?
# df.index
df.head()
```

```python
f = df.pivot(index='date', columns='city')
f
```

```python
# values mean what you want to your value be
f= df.pivot(index='date', columns='city', values='humidity')
f
```

```python
df.pivot(index='humidity', columns='city')
```

# Pivote table 
#### is used to summarize and aggregate data inside dataframe

```python
df = pd.read_csv('weather2.csv')
df
```

```python
df.pivot_table?
```

```python
df.pivot_table(index='city', columns='date') #defaul aggfunc is mean
```

```python
df.pivot_table(index='city', columns='date', aggfunc='sum')
```

```python
df.pivot_table(index='city', columns='date', aggfunc='sum', margins=True)# if margin is true you have all and is mean of two number
```

### grouper in pivot table

```python
df = pd.read_csv('weather3.csv')
```

```python
df
```

```python
# when you read csv date is not realy a date it's string obj you must to convert date 
df['date'] = pd.to_datetime(df['date'])
```

```python
# you can apply grouper fun to aggregate based on date frequency 
df.pivot_table(index=pd.Grouper(freq='M', key='date'), columns='city') # freq = M it's means frequency of month
```
